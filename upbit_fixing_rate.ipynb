```python
import openpyxl
import requests
from datetime import timedelta
import time
import json
```

```python
filename = "fixing_rate/3.1 바이낸스 계정 횡령 외 UST 인출, 현금 인출 관련 정리자료_v3_2.xlsx"
workbook = openpyxl.load_workbook(filename=filename, data_only=True)
sheet = workbook['Sheet2']
```

```python
headers = {"Accept": "application/json"}
for i in range(2, sheet.max_row + 1):
    date_time = sheet[f"A{str(i)}"].value
    utc_time = date_time + timedelta(hours=-9)
    market = sheet[f"E{i}"].value
    if market == "KRW":
        continue
    else:
        market = "KRW-"+market

    url = f"https://api.upbit.com/v1/candles/minutes/1?market={market}&to={utc_time.strftime('%Y-%m-%d %H:%M:%S')}&count=1"
    try:
        res = requests.get(url, headers=headers)
        time.sleep(0.1)
        data =json.loads(res.text)
        trade_price = data[0]["trade_price"]
        sheet[f"G{i}"].value = trade_price
        print(utc_time, market, trade_price)
    except Exception as e:
        print(e)
        print(f"{market} 상장폐지")
        
```

```python
workbook.save("fixing_rate/try1.xlsx")
```
