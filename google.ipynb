```python
import openpyxl
filename = 'files/Bbrick_취합파일 정리_2200623 (입금완료).xlsx'
excel_wb = openpyxl.load_workbook(filename=filename, data_only=True)
excel_ws = excel_wb['입금']
```

```python
ex_from_address_column = excel_ws['N'][1:]
ex_from_memo_column = excel_ws['O'][1:]
ex_from_address_list = list(map(lambda x: x.value ,ex_from_address_column))
ex_from_memo_list = list(map(lambda x: x.value if x.value is not None else "" ,ex_from_memo_column))
```

```python
ex_from_address = ex_from_address_list[0]
ex_from_memo = ex_from_memo_list[0]
print(ex_from_address, ex_from_memo)
```

```python
hash_map = {}
for i in range(excel_ws.max_row - 1 ):
    ex_from_address = ex_from_address_list[i]
    ex_from_memo = ex_from_memo_list[i]
      
    

    
    if ex_from_address == None:
        continue

    # 거래소 지갑 찾기
    if "(" in ex_from_address:
        first_space = ex_from_address.find(" ")

        exchange_address = ex_from_address[:first_space]
        exchange_name = ex_from_address[first_space + 2: -1]

        if (exchange_address, ex_from_memo) in gs_address_memo_list:
            # 아무것도 안들어와
            
            # 메모가 존재하면


            if exchange_address + " " + ex_from_memo in hash_map:
                continue
            else:
                hash_map[exchange_address + " " + ex_from_memo] = exchange_name


            print(f"exchange matched: row {str(i+2)} exchange_address: {exchange_address}, exchange_name: {exchange_name}, memo: {ex_from_memo}")
```
